```python
### Assignment 13: Nested Tuple Iteration
# Create a nested tuple and iterate over the elements, printing each element.
nested_tuple = (1,32,3,(2,4,'s','thr'), 'sd','asd','assd')
def print_elements(data: tuple) -> None:
    for elem in data:
        if isinstance(elem, tuple):
            print_elements(elem)  # recursive call
        else:
            print(elem, end=", ")


print_elements(nested_tuple)

# for elem in nested_tuple:
#     if isinstance(elem, tuple):
#         for sub in elem:
#             print(sub, end=" ")
#     else:
#         print(elem, end=" ")
```
