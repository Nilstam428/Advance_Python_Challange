```python
with open('example.txt', mode='r') as file:
    content = file.read()
    print(content)
```

```python

with open('example.txt', mode='r') as file:
    for line in file:
        print(line.strip()) # remove the newline character
```

```python
# writing/overwriting a file
with open('example.txt', mode='a') as file:
    file.write('This is a new line.\n')
```
