```python
# Problem 1 — Frequency Counter
data = ["apple", "banana", "apple", "orange", "banana", "apple"]
counts = {}
for item in data:
  if item in counts:
    counts[item] += 1
  else:
    counts[item] = 1
  
counts
```

```python
# Problem 2 — Remove Duplicates (Keep Order)
data = [1, 2, 2, 3, 4, 3, 5]
result = []
for item in data:
  if item not in result:
    result.append(item)
print(data)
print(result)
```

```python
# Problem 3 — Find Second Largest Number
data = [10, 20, 4, 45, 99]
data.sort()
print(data[-2])
```

```python
# Problem 4 — Flatten a Nested List
data = [[1, 2], [3, 4], [5]]
li = []
for item in data:
    for values in item:
        li.append(values)

print(li)
```

```python
# Problem 5 — Word Frequency in Sentence
text = "python is good python is easy"
li = text.join
frequency = {item: text.count(item) for item in li}
print(frequency)    
```
