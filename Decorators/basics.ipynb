```python
# Basic python decorator code
from functools import wraps

def my_decorator(func):
  @wraps(func)
  def wrapper():
    print("Before function calls")
    func()
    print("After function calls")
  return wrapper


@my_decorator
def greet():
  print("Hello from decorator class:")
  
greet()
# print(greet.__name__)
```

```python
from functools import wraps

def logger(func):
  @wraps(func)
  def wrapper(*args, **kwargs):
    print(f"Calling: {func.__name__}")
    result = func(*args, **kwargs)
    print(f"Finished: {func.__name__}")
    return result
  return wrapper

@logger
def add(x, y):
  return x + y

@logger
def subtract(x, y):
  return x - y

result = add(2, 3)
result1 = subtract(2, 3)
print(result)
print(result1)
```
