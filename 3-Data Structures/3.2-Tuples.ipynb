```python
my_tuple = (1,2,3,4,5)
print(my_tuple)
print(type(my_tuple))
```

```python
lst = list()
print(type(lst))
tup = tuple()
print(type(tup))
```

```python
## tuples are immutable
print(my_tuple)
try:
  my_tuple[0] = 10
except Exception as e:
  print(e)
print(my_tuple)
```

```python
# declaration of tuple
tup = (1,2,3,4,5)
# access the elements of tuple
print(tup[0])
print(tup[-1])
print(tup[:4])
print(tup[::2])
```

```python
# packing and unpacking of tuple
packed_tuple = 1,"helllo",True,3,4
print(packed_tuple)
```

```python
# unpacking tuple
a,b,c,d,e = packed_tuple
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
# unpacking with * operator
a, *middle, b = packed_tuple
print(a)
print(middle)
print(b)
```

```python
# nested tuple
nested_tuple = (1,2,3),(4,5,6),(7,8,9)
# access 5
print(nested_tuple[1][1])
```

```python
# iterating over nested tuples
for sub_tuple in nested_tuple:
  # print(sub_tuple)
  for element in sub_tuple:
    print(element,end=" ")
  print()
```
