### Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements.
 
#### Sets are useful for: 
- membership tests, 
- eliminating duplicate entries
- performing mathematical set operations 
  - like union, 
  - intersection
  - difference
  - symmetric difference.

```python
## sets are created using {}
set1 = {1,2,3,4,5}
print(set1)
print(type(set1))
```

```python
empty_set = set()
print(empty_set)
print(type(empty_set))
```

```python
my_set=set([1,2,3,4,5,6])
print(my_set)
print(type(my_set))
```

```python
# basic set operations
my_set.add(7)
print(my_set)
my_set.remove(2)
print(my_set)
```

```python
# remove - removes an element from the set and raise error if the element is not present
# discard - removes an element from the set if present does, and not raise error if not present
# pop - removes and returns an element from the set
# clear - removes all elements from the set

my_set.pop()
print(my_set)
my_set.clear()
print(my_set)
```

```python
# membership test
my_set = {1,2,3,4,5}
print(1 in my_set)
print(6 in my_set)
print(6 not in my_set)
```

```python
## mathematical operations
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set2.difference(set1))
print(set1.symmetric_difference(set2))
print(set1.issubset(set2))
print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set2.issuperset(set1))
```

```python
# counting unique word in the test 
my_text = "This is a practice tutorial i am trying this is amazing"
words = my_text.split()
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
