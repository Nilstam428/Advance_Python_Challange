```python
# yield generates the data for a function in a class format
def get_detail():
  yield "Nilesh"
  yield "Tamboli"
  yield "23"
  
user = get_detail()
```

```python
for details in user:
  print(details)
```

```python
def get_detail_specific():
  yield "first_name: Nilesh"
  yield "last_name: Tamboli"
  yield "age: 23"
  
```

```python
user1 = get_detail_specific()
print(user1)
print(next(user1))
print(next(user1))
print(next(user1))
```

```python
def infinite_chai():
    count = 1
    while True:
        yield f"Refil #{count}"
        count += 1

refill = infinite_chai()
user2 = infinite_chai()

for _ in range(5):
    print(next(refill))

for _ in range(6):
    print(next(user2))
```

```python
def chai_customer():
    print("Welcome ! What chai would you like ?")
    order = yield
    while True:
        print(f"Preparing: {order}")
        order = yield

stall = chai_customer()
next(stall) # start the generator

stall.send("Masala Chai")
stall.send("Lemon Chai")
```

```python
# clear the contents of the generator from the memory
stall.close()
```
