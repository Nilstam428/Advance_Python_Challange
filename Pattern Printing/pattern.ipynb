```python
# Increasing pattern from lleft to right
num = 5
for i in range(num):
    for j in range(i+1):
        print("*", end="")
    print()
```

```python
# Decreasing Pattern from from left to right
for i in range(num):
    for j in range(i,num):
        print("*", end="")
    print()
```

```python
# Increasing pattern from right to left 
# for this we need an empty Decreasing pattern of empty space and then incrreasing pattern
for i in range(num):
    for j in range(i,num):
        print(" ",end="")
    
    for j in range(i+1):
        print("*", end="")
    
    print()
```

```python
# Decreasing pattern from right to left
for i in range(num):
    for j in range(i+1):
        print(" ", end="")

    for j in range(i,num):
        print("*", end="")
    print()
```

```python
# pyramid pattern 
n = 6

for i in range(1, n + 1):

    # spaces
    for j in range(i, n):
        print(" ", end="")

    # stars
    for k in range(1, 2*i):
        print("*", end="")

    print()
```
