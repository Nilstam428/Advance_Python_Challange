```python
# Left Half pyramid
num = int(input("Enter no. rows"))
for i in range(num+1):
    print("*"*i)
    
```

```python
for i in range(1,num+1):

    for j in range(1,i+1):
        print("*",end="")
    print()
```

```python

for i in range(1,num+1):
    for j in range(1,num-i+2):
        print("*", end="")
    print()
```

```python
for i in range(1,num+1):
    for i in range(i, num+1):
        print(i, end="")
    print()
```

```python

for i in range(num):
    for j in range(i+1):
        print("*", end="")
    print()
```

```python
for i in range(num):
    for j in range(i,num):
        print("*", end="")
    print()
```

```python
for i in range(num):
    for j in range(i,num):
        print(" ", end="")
    
    for j in range(i+1):
        print("*", end='')

    print()
```

```python
for i in range(num):
    for j in range(i+1):
        print(" ", end="")

    for j in range(i,num):
        print("*", end="")
    print()
```

```python
for i in range(num):
    for j in range(i,num):
        print(" ", end="")
    for j in range(i):
        print("*", end="")
    for j in range(i+1):
        print("*", end="")
    print()
```
