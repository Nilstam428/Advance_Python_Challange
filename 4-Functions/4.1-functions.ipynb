### Functions in Python
##### A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

- Function Parameters

```python
def show():
  print('Hello Nilesh! ')
```

```python
show()
```

```python
## function parameters
def greet(name):
  print(f"Hello {name}! How are you?")
name = input("Enter your name.")
greet(name)
```

```python
# Default parameter
def greet(your_name='Guest'):
  print(f"Hello {your_name}! How are you.")
greet()
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,3,5,3,23,)
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
# takes input as key value pair - dictionary
print_details(name='rahul', age=34, work='student')
```

```python
def print_details(*args,**kwargs):
    for val in args:
      print(f"positional argument: {val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
# takes input as key value pair - dictionary
print_details(12,3,4,5,name='rahul', age=34, work='student')
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```
