### Encapsulation
#### Encapsulation is the process of hiding the implementation of a class from the user. It is used in the case of sensitive data.


```python
# Access Modifier
# 1.Public - default variable
# 2.Private - __variable Name
# 3.Protected - _variable Name

class Person:
  def __init__(self, name, age, gender, address, phone_no):
    self.name = name 
    self.age = age
    self.__gender = gender
    self._address = address
    self.__phone_no = phone_no

  def get_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Gender: {self.__gender}")
    print(f"Address: {self._address}")
    print(f"Phone No: {self.__phone_no}")

  def set_info(self, name, age, gender, address, phone_no):
    self.name = name 
    self.age = age
    self.__gender = gender
    self._address = address
    self.__phone_no = phone_no
# either i set values directly or i can use set_info
Person1 = Person('Nilesh',22,'Male','Pune','1234567890')
# using set_info for upadte or change the details
Person2 = Person('Suresh',25,'Male','Pune','1234867890')
# Person2.set_info('Suresh',25,'Male','Pune','1234867890')
print("Person 1 details:")
print(Person1.get_info())

print('===================================================================')
print("Person 2 details:")
print(Person2.get_info())

# print(Person1._address)


```

```python
# Using set_info for updating person2 details
Person2.set_info('Suresh',35,'Male','Pune','12449847890')
print(Person2.get_info())
```
